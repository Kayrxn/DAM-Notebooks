## OBJETOS INMUTABLES


```python

a = 3
b = a               # b es 3 -- porque ahora B apunta al valor de A en memoria
print("b es ", b)

a = 3               # a vuelve a ser 3, pero eso no cambiaba
b = a               # y B vuelve a referir al valor de A, que sigue siendo 3, o sea, se queda igual.
a = 'spam'          # El valor de A cambia al string "spam".
print("b es ", b)   # printea que B es 3 porque es el valor al que apuntaba hasta ahora. Eso no cambia.
```

Los objetos `integer` no son mutables:


```python
a = 3
b = a                   
a = a + 2               # a cambia a 3 + 2 que es 5. B no cambia en ningn momento porque ya apuntaba al valor "3" en memoria.
print("b es ", b)       
```

## OBJETOS MUTABLES

### Listas

```python

L = [2, 3, 4]	    # L = 2, 3, 4
M = L			    # M apunta a la misma lista de valores que L en memoria: "2, 3, 4"
L[0] = 24 		    # El primer "puesto" del valor de la lista, pasa a ser 24: "24, 3, 4"

print("L = ", L)    # Por lo tanto, tanto L como M se refieren a la misma lista, que ahora tiene los valores 23,3,4
print("M = ", M)   
```

Copia de objetos en vez de referencia:


```python

L = [2, 3, 4]
M = L[:]		    # Es diferente a M = L, el operador [:] crea una copia de los elementos en otra lista diferente.
L[0] = 24

print("L = ", L)    # Por lo tanto, L cambia a 24,3,4

print("M = ", M)    # Y M se queda con su copia a parte de 2,3,4
```

### Shared References and Equality
- Operador `==` igualdad de valores.
- Operador `is` identidad de objetos.

```python

L = [1, 2, 3]       # L es la lista de: 1,2,3
M = L 			    # M apunta al mismo lugar que L: la lista 1,2,3.

L == M 			    # Aqu comprueba si L <tiene el mismo valor> que M
L is M 			    # Aqu comprueba si L <apunta> a la misma referencia en memoria que M.
```

```python
L = [1, 2, 3]
M = [1, 2, 3]	   # M y L apuntan a listas con los mismos valores, pero son diferentes listas para cada uno. 

print(
    "L == M",       # M tiene el mismo valor que L? S.
    L == M 			
)

print(
    "L is M", 	    # M apunta al mismo objeto en memoria? No. Son listas distintas.
    L is M
)
```

#### CACHE

Because small integers and strings are cached and reused, though, is tells us they reference the same single object.


```python

X = 42
Y = 42		    

X == Y         # X tiene el mismo valor que Y? S.
X is Y 		   # X apunta al mismo objeto en memoria? S, 42 es 42, es el mismo nmero entero.
```

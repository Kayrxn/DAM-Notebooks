```python
# Define el dibujo de la línea separadora
sep = '-' * 45 + '\n'
sep2 = '\n' + '-' * 45
```

```python
print(sep + 'EXCEPTION RAISED AND CAUGHT' + sep2)

try:
 x = 'spam'[99]

except IndexError:
 print('except run')

finally:
 print('finally run')

print('after run')
```

 ####  <span style="color:pink">✨EXPLICACION</span>  
Intenta hacer `'spam'[99]` pero spam no tiene 99 posiciones, así que tira un error.  
Como suelta error, ejecuta except.  
finally siempre se ejecuta.  

```python
print(sep + 'NO EXCEPTION RAISED' + sep2)

try:
 x = 'spam'[3]

except IndexError:
 print('except run')

finally:
 print('finally run')
 
print('after run')
```

 ####  <span style="color:pink">✨EXPLICACION</span>  
Intenta hacer `'spam'[3]` y es correcto porque spam tiene 4 posiciones, así que no tira indexerror.  
finally siempre se ejecuta.

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE' + sep2)

try:
 x = 'spam'[3]

except IndexError:
 print('except run')

else:
  print('else run')

finally:
 print('finally run')
 
print('after run')
```

 ####  <span style="color:pink">✨EXPLICACION</span>  
Intenta hacer el `'spam'[3]`, y es correcto, por lo tanto ignora except.  
Como no suelta error, ejecuta el else.  
finally siempre se ejecuta.

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT' + sep2)

try:
 x = 1 / 0

except IndexError:
 print('except run')

finally:
 print('finally run')
 
print('after run')
```

 ####  <span style="color:pink">✨EXPLICACION</span>  
No puede dividir 1 entre 0, así que lanza error.  
Se ejecutaría el except si usáramos el error correcto, `ZeroDivisionErrror`. Pero no, aquí busca IndexErrror.  
Y finalmente, finally.

```python
try:
    x = 0
    y = 10 / x
except ValueError:
    print("Debes ingresar un número válido.")
except ZeroDivisionError:
    print("No puedes dividir por cero.")
else:
    print("Todo salió bien:", y)
finally:
    print("Fin del programa (bloque finally).")
```

 ####  <span style="color:pink">✨EXTRA</span>
Primero comprueba si es un número entero. 
Después comprueba si es un 0.  
Si no pasó ninguno, "todo salió bien" y ejecuta la división.
